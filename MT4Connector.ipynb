```python
vol_window = 30
obs_window = 20
criteria = [1,5]
forward = 30
ref = SPX
```

```python
Pair = Cable
Pair['ref_close']=ref['close']
Pair['ref_vol']= ref['volume']
Pair = Pair.drop(['spread','real_volume'], axis= 1)
trainidx = Pair.index[0:int(len(Pair.index)*2/2)]
Pair_train=Pair.loc[trainidx]
Pair_train['ret']=Pair_train['close'].pct_change()*100
Pair_train['volatility']= Pair_train['close'].rolling(vol_window).std()
Pair_train['skewness']= Pair_train['close'].rolling(vol_window).skew()
Pair_train['kurtosis']= Pair_train['close'].rolling(vol_window).kurt()
Pair_train['period_low'] = Pair_train['low'].rolling(obs_window).min()
Pair_train['period_high'] = Pair_train['high'].rolling(obs_window).max()
trendidx = (Pair_train['close'] - Pair_train['low'])>(Pair_train['high'] - Pair_train['close'])
Pair_train['U_trend']= trendidx
Pair_train['multiple']= (Pair_train['close']-Pair_train['period_high'])/Pair_train['volatility']
Pair_train.loc[trendidx,'multiple'] = (Pair_train[trendidx]['close']-Pair_train[trendidx]['period_low'])/Pair_train[trendidx]['volatility']
Pair_train['future_pip_move'] = (Pair_train['close'].shift(-forward)-Pair_train['close'])*100000
Pair_train['Mom_Rev'] = abs(Pair_train['future_pip_move']) > 200000*Pair_train['volatility']

mean_vol = Pair_train['volume'].mean()
Pair_train=Pair_train.dropna()
Pair_Chart = Pair_train[Pair_train['volume']> mean_vol]


def evaluate(criteria):
    ins = (Pair_train['multiple']>=criteria[0]).astype(int)* (Pair_train['multiple']<criteria[1]).astype(int)
    In_group = Pair_train[ins.astype(bool)]
    Out_group = Pair_train[-ins.astype(bool)]
    mean_diff = In_group['future_pip_move'].mean()- Out_group['future_pip_move'].mean()
    return mean_diff
```

```python
Criteria_range =np.array(range(0,20))/2
n = len(Criteria_range)-1
diffs = np.zeros(shape = (n,n))
for i,c in enumerate(Criteria_range[1:(n+1)]):
    for j,d in enumerate(Criteria_range[0:(i-1)]):
        cri = [c,d]
        diffs[(i-1),j]=evaluate(cri)
```
